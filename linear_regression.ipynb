```python
import pandas as pd 
#it is one of python lib .pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool
```

```python
dataset=pd.read_csv('Salary_Data.csv') # csv file read here
```

```python
dataset # display  data from specify file
```

```python
independent =dataset[["YearsExperience"]] #input variable
independent
```

```python
dependent=dataset[["Salary"]] #output variable
dependent
```

```python
from sklearn.model_selection import train_test_split #sklearn model used in machine learning and this import command split the train and test sample
```

```python
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
x_train # input train sample
```

```python
y_train
#output train sample
```

```python
from sklearn.linear_model import LinearRegression #in this model used to linear regression import from sklearn 
```

```python
regressor=LinearRegression() #call linear regression function
regressor.fit(x_train,y_train) 
# here create model ,its learn everything and find weight(m) and bias(intercept) for example model:y=mx+b
#in this equation check prediction value
```

```python
weight=regressor.coef_ # in this value possible change slope so its desired to data is linear ir not 
weight
```

```python
bais=regressor.intercept_ #In this value is orgin of the linear so if you want start or orgin value
bais
```

```python
y_pred=regressor.predict(x_test) #it is predict the train test sample
```

```python
from sklearn.metrics import r2_score # in this model evaluating between test data and predict value ,itwill check the model how much learn
r_score=r2_score(y_test,y_pred) #here check dependent actual value and predict value if the answer will get near to 1 is good learn model otherwise it is bad .r2 range between 0-1
```

```python
r_score # r2  value near to 1 so in this model is good
```

```python
import pickle # Python pickle module is used for serializing and de-serializing a Python .
```

```python
filename="finalized_model_linear.sav" #create file
```

```python
pickle.dump(regressor,open(filename,"wb")) # in this dump command copy the data from regressor(include weight,bais),wb=write binary rb=read 
```

```python
loaded_model=pickle.load(open("finalized_model_linear.sav","rb")) #take pickle model load the file and read binary
result=loaded_model.predict([[13]])# check predict value
```
